```python
# train test splits
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# change single dimension to 2D
xtrain=xtrain.values.reshape(-1,1) # row,col -1 unknown
```

```python
xtest=xtest.values.reshape(-1,1)
ytest=ytest.values.reshape(-1,1)
ytrain=ytrain.values.reshape(-1,1)
```

```python
xtest.shape
```

model  building

simple linear regression algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
# fit
linreg.fit(xtrain,ytrain)
```

```python
print(linreg.intercept_)
```

```python
# m slope
linreg.coef_
```

```python
ytest # actual data -  dataset
```

```python
ypred = linreg.predict(xtest) # predicted data
print(ypred)
```

```python
# residual error
err=ytest - ypred
err
```

```python
plt.scatter(xtest,ytest,color='blue') # actual data
plt.plot(xtest,ypred,color='red') # predicted
plt.title('Height Weight prediction')
```

```python
# Evaluation metrics
from sklearn import  metrics
print("Mean absolute error MAE ",metrics.mean_absolute_error(ytest,ypred))
```

```python
print("Mean squared error MSE ",metrics.mean_squared_error(ytest,ypred))
```

```python
print("Root mean squared error ",np.sqrt(metrics.mean_squared_error(ytest,ypred)))
```

```python
print("R2 score ",metrics.r2_score(ytest,ypred))
# near to one, good model
```
